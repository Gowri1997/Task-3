which is the most and least profitted city in united states?

```python
superstore_sample_1=superstore_sample.groupby(['City'],as_index=False).sum().sort_values(by='Profit',ascending=False)
superstore_sample_1
```

```python
superstore_sample_1[superstore_sample_1.Profit==superstore_sample_1.Profit.max()].City
```

```python
superstore_sample_2=superstore_sample.groupby(['Category'],as_index=False).sum().sort_values(by='Profit',ascending=False)
superstore_sample_2
```

```python
plt.figure(figsize=(10,7))
X=superstore_sample_2.Category

Y_sales=superstore_sample_2['Sales']

Y_Profit=superstore_sample_2['Profit']

X_axis=np.arange(len(superstore_sample_2.Category))

plt.bar(X_axis-0.1,Y_sales,0.2,label='Sales')

plt.bar(X_axis+0.1,Y_Profit,0.2,label='Profit')

plt.xticks(X_axis,X)
plt.ylabel('Sales and Profit')
plt.legend()
plt.show()
```

```python
superstore_sample_3=superstore_sample.groupby(['Sub-Category'],as_index=False).sum().sort_values(by='Profit',ascending=False)
superstore_sample_3
```

```python
plt.figure(figsize=(10,7))
plt.plot(superstore_sample_3['Sub-Category'],superstore_sample_3.Sales,color='g',label='sales')
plt.plot(superstore_sample_3['Sub-Category'],superstore_sample_3.Profit,color='r',label='profit')
plt.xticks(rotation=90)
plt.title('Sum of Sales and Sum of Profit by SubCategory')
plt.ylabel("Sum of Sales and Sum of Profit")
plt.xlabel('SubCategory')
plt.legend()
plt.show()
```

From this we can identify somecategory of items whose sales increase but the profit decreases in anomolous .This will give us insight into the demand of the item
